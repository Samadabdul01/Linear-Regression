```python
import numpy as np 
from sklearn import datasets, linear_model
diabetes = datasets.load_diabetes()
```

```python
diabetes.data
```

```python
diabetes.keys()
```

```python
diabetes.DESCR
```

```python
diabetes_X = diabetes.data
```

```python
diabetes_X
```

```python
diabetes_X_train = diabetes_X[:-30]
diabetes_X_test = diabetes_X[-30:]
diabetes_y_train = diabetes.target[:-30]
diabetes_y_test = diabetes.target[-30:]
```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_X_train,diabetes_y_train)
```

```python
model.predict(diabetes_X_test)
```

```python
diabetes_y_predict = model.predict(diabetes_X_test)
```

```python
from sklearn.metrics import  mean_squared_error
print('Mean squared error:',mean_squared_error(diabetes_y_test,diabetes_y_predict))
```

```python
print('Weights:',model.coef_)
print('Intercept:',model.intercept_)
```

Linear Regression on Iris Dataset

```python
from sklearn import  datasets,linear_model
import numpy as np
iris = datasets.load_iris()
```

```python
iris.data
```

```python
iris.keys()
```

```python
iris_X = iris.data
```

```python
iris_X
```

```python
iris_X_train = iris_X[:-30]
iris_X_test = iris_X[-20:]
iris_y_train = iris.target[:-30]
iris_y_test = iris.target[-20:]
```

```python
model = linear_model.LinearRegression()
model.fit(iris_X_train,iris_y_train)
```

```python
model.predict(iris_X_test)
```

```python
iris_y_predict = model.predict(iris_X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(iris_y_test,iris_y_predict)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(iris_y_predict,iris_y_test)
plt.plot(iris_y_predict,iris_y_test)
```

```python
import seaborn as sns
sns.regplot(x = iris_y_predict, y = iris_y_test, data = iris_X)
plt.show()
```
